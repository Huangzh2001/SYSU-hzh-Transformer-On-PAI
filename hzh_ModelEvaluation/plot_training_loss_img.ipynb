0.导入库

```python
import numpy as np
import os
# 导入库
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
```

1.在服务器里面训练完以后的loss_list存放的路径

```python
loss_path = '../hzh_SwinUnet/loss'
train_loss_path = loss_path + '/' + 'train_loss_list.npy'
val_loss_path = loss_path + '/' + 'val_loss_list.npy'
```

2.读取loss

```python
train_loss = np.load(train_loss_path).tolist()
val_loss = np.load(val_loss_path).tolist()
print(len(train_loss))
```

3.画loss变化折线图

```python
# 生成epoch坐标轴
epoch_num = len(train_loss)
x_label = []
for i in range(epoch_num):
    x_label.append(i+1)
print(x_label)
```

```python
# 画图
figure=plt.figure(figsize=(11,5),dpi=80)
 
# 子图1
ax1=plt.subplot(121)
train_loss_line = plt.plot(x_label, train_loss,linestyle=':',marker='o')
plt.xlabel("epoch")
plt.ylabel("train loss")
 
# 子图2
ax2=plt.subplot(122)
val_loss_line = plt.plot(x_label, train_loss,linestyle=':',marker='o')
ax2.set_xlabel("epoch") 
plt.ylabel("val loss")

figure.suptitle('Train and Val Loss of each epoch')

plt.tight_layout()
plt.savefig("../Image/loss.png")
plt.show()
```

```python
from skimage.metrics import structural_similarity as ssim
from PIL import Image
import numpy as np

import os, sys
```

下面画论文要用的图

1.先初始化要用的库

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import data, img_as_float
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import mean_squared_error
```

2.读取示例图片

```python
i = 2
ori_img = np.array(Image.open(os.path.join(pre_image_path,rec_50_sensor_dir[i])))
rec_50_img = np.array(Image.open(os.path.join(rec_50_sensor_path,rec_50_sensor_dir[i])))
rec_100_img = np.array(Image.open(os.path.join(rec_100_sensor_path,rec_50_sensor_dir[i])))
rec_200_img = np.array(Image.open(os.path.join(rec_200_sensor_path,rec_50_sensor_dir[i])))
rec_400_img = np.array(Image.open(os.path.join(rec_400_sensor_path,rec_50_sensor_dir[i])))
rec_800_img = np.array(Image.open(os.path.join(rec_800_sensor_path,rec_50_sensor_dir[i])))

rows, cols = rec_50_img.shape
```

4.开始画图

```python
# 先计算各图与原图的mse、ssim
## 1.计算mse
mse_self = mean_squared_error(ori_img, ori_img)
mse_50 = mean_squared_error(rec_50_img, ori_img)
mse_100 = mean_squared_error(rec_100_img, ori_img)
mse_200 = mean_squared_error(rec_200_img, ori_img)
mse_400 = mean_squared_error(rec_400_img, ori_img)
mse_800 = mean_squared_error(rec_800_img, ori_img)

## 2.计算SSIM
ssim_self = ssim(ori_img, ori_img, multichannel=True)
ssim_50 = ssim(rec_50_img, ori_img, multichannel=True)
ssim_100 = ssim(rec_100_img, ori_img, multichannel=True)
ssim_200 = ssim(rec_200_img, ori_img, multichannel=True)
ssim_400 = ssim(rec_400_img, ori_img, multichannel=True)
ssim_800 = ssim(rec_800_img, ori_img, multichannel=True)
```

```python
# 计算Psnr。这时要调用库
from skimage.metrics import peak_signal_noise_ratio as psnr
psnr_self = psnr(ori_img, ori_img)
psnr_50 = psnr(rec_50_img, ori_img)
psnr_100 = psnr(rec_100_img, ori_img)
psnr_200 = psnr(rec_200_img, ori_img)
psnr_400 = psnr(rec_400_img, ori_img)
psnr_800 = psnr(rec_800_img, ori_img)
```

```python
# 初始化matplotlib并画图
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(9, 6), sharex=True, sharey=True)
ax = axes.ravel()
label = 'MSE: {:.2f}, SSIM: {:.2f}, PSNR: {:.2f}'

ax[0].imshow(ori_img, cmap=plt.cm.gray, vmin=0, vmax=255)
ax[0].set_xlabel(label.format(mse_self, ssim_self, psnr_self))
ax[0].set_title('Original image')

ax[1].imshow(rec_50_img, cmap=plt.cm.gray, vmin=0, vmax=255)
ax[1].set_xlabel(label.format(mse_50, ssim_50, psnr_50))
ax[1].set_title('image of 50 sensors')

ax[2].imshow(rec_100_img, cmap=plt.cm.gray, vmin=0, vmax=255)
ax[2].set_xlabel(label.format(mse_100, ssim_100, psnr_100))
ax[2].set_title('image of 100 sensors')

ax[3].imshow(rec_200_img, cmap=plt.cm.gray, vmin=0, vmax=255)
ax[3].set_xlabel(label.format(mse_200, ssim_200, psnr_200))
ax[3].set_title('image of 200 sensors')

ax[4].imshow(rec_400_img, cmap=plt.cm.gray, vmin=0, vmax=255)
ax[4].set_xlabel(label.format(mse_400, ssim_400, psnr_400))
ax[4].set_title('image of 400 sensors')

ax[5].imshow(rec_800_img, cmap=plt.cm.gray, vmin=0, vmax=255)
ax[5].set_xlabel(label.format(mse_800, ssim_800, psnr_800))
ax[5].set_title('image of 800 sensors')

plt.tight_layout()
# 将图片保存
plt.savefig('D:\\Document\\NutCloud\\中山大学文件\\2023年寒假\\毕业论文\\论文图片\\test_6.png')
plt.show()
```
